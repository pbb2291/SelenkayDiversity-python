```python
import pandas as pd
import pygam
import matplotlib.pyplot as plt
import numpy as np

# Open outputs from BirdSurveyCleaning.ipynb on Google drive (peter.brehm.boucher folder)
SpotSppDF = pd.read_csv('./data/in/SelenkayBirds_SpotSppDF.csv')
SpotSummaryStats = pd.read_csv('./data/in/DFBirds_SummaryStatsbySpot.csv')
```

```python
SpotSppDF.head()
```

```python
# Add an abundance col (literally, number of raw counts)
# Note- dropping unnamed col for the calculation
# but keeping it to join with Summary stats df 
SpotSppDF['Abundance'] = SpotSppDF.drop('Unnamed: 0', axis=1).sum(axis=1)
```

```python
SpotSppDF.Abundance.head()
```

```python
SpotSummaryStats.head()
```

```python
# Add the dfs together
SpotSummaryStats_final = SpotSummaryStats.merge(SpotSppDF[['Unnamed: 0', 'Abundance']])
SpotSummaryStats_final.drop('Unnamed: 0', axis=1, inplace=True)
SpotSummaryStats_final.head()
```

```python
SpotSummaryStats_final.to_csv('./data/out/SpotSummaryStats_Birds_final.csv', index=False, header=True)
```

```python
# Also, clean and export a version of SpotSppDF
# Add in Spots
SpotSppDF_final = SpotSppDF.merge(SpotSummaryStats[['Unnamed: 0', 'Spot']])
SpotSppDF_final.drop(['Unnamed: 0', 'Abundance'], axis=1, inplace=True)
SpotSppDF_final.set_index('Spot', inplace=True)
SpotSppDF_final.head()
```

```python
# Export
SpotSummaryStats_final.to_csv('./data/out/SpotSppDF_Birds_final.csv',
                              index=True, header=True)
```
